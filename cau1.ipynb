```python
dff['PASS 6/2020?'] = np.where(dff['Doanh số 6/2020']>dff['Target 6/2020'],'PASS','NOT')
dff['PASS 5/2020?'] = np.where(dff['Doanh số 5/2020']>dff['Target 5/2020'],'PASS','NOT')
dff['PASS 4/2020?'] = np.where(dff['Doanh số 4/2020']>dff['Target 4/2020'],'PASS','NOT')
```

```python
dfff = dff[['PASS 6/2020?','PASS 5/2020?','PASS 4/2020?']]
```

```python
dfff
```

```python
dffff = dfff.apply(pd.value_counts, axis=1).fillna(0)
```

```python
dffff
```

```python
dfffff = dff[['PIC']]
```

```python
a = dfffff.merge(dffff, left_index=True, right_index=True)
```

```python
a
```

```python
b = a.groupby('PIC').sum()
```

```python
b
```

```python
b['%PASS'] = round((b.PASS / (b.NOT + b.PASS))*100, 2)
```
